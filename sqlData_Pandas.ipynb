```python
data = None
with open("jsonData.json" , 'r') as dataFile:
  data = json.load(dataFile)

print('read')
```

```python
data['data']
```

### empty dataframe

```python
df = pd.DataFrame(columns = ['age' , 'id' , 'name' , 'score'])
df
```

```python
%%time
# append list of items dictionary and assign it to df
for index , item in enumerate(data['data']):
  print(f"-- running for {index} --")
  df = df.append([item] , ignore_index = True)

print('\n\n done')
```
